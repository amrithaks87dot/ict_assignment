```python
#1
def find_max(a, b, c):
  if a >= b and a >= c:
    return a
  elif b >= a and b >= c:
    return b
  else:
    return c
num1 = 10
num2 = 25
num3 = 15

maximum = find_max(num1, num2, num3)
print(f"The maximum of {num1}, {num2}, and {num3} is: {maximum}")
```
